```python
from plottr.data.datadict_storage import datadict_from_hdf5
import numpy as np
import matplotlib.pyplot as plt
import lmfit
```

```python
import sklearn
from sklearn.decomposition import PCA
```

```python
def Lorentzian(freq, amp, const, center, k):
    return const - amp*k**2/((freq-center)**2 + k**2)
```

```python
header = "D:/K_sunada/result/CDY161/"
# twotone(redout: 10.26GHz)
three_peak = "/2024-07-06/2024-07-06T121306_aa55295a-cw_twotone" #drive: R.T.-40dB, lo power: -20~24dB, readout: R.T.: -30 dB, vna power: -10dB
ge_peak ="/2024-07-06/2024-07-06T122144_df02b973-cw_twotone"#<ge>drive: R.T.-60dB, lo power: -20~10dB, readout: R.T.: -30 dB, vna power: -10dB
ef_peak = "/2024-07-06/2024-07-06T121712_3c93ccad-cw_twotone" #<ef>drive: R.T.-40dB, lo power: 0~24dB, readout: R.T.: -30 dB, vna power: -10dB
```

ge, ef peak

```python
ddge = datadict_from_hdf5(header + ge_peak +"/data")
ddef = datadict_from_hdf5(header + ef_peak +"/data")
```

```python
def PCA_transform(Re_c, Im_c):
    pca = PCA()
    c_data = np.concatenate([np.array([Re_c]).T, np.array([Im_c]).T], axis = 1)
    pca.fit(c_data)
    c_pca = pca.transform(c_data)
    c_PCA1 = c_pca.T[0]
    c_PCA2 = c_pca.T[1]
    return c_PCA1, c_PCA2
```

## ge遷移
初期値設定

```python
S11ge = ddge['s11']['values']
ge_phase = np.angle(S11ge)
freqge = ddge['frequency']['values']

y = ge_phase[3]
x = freqge[3]
plt.plot(x, y)
plt.show()
```

```python
I = (S11ge.real)[3]
Q = (S11ge.imag)[3]
#c_data = np.concatenate([np.array([I]).T, np.array([Q]).T], axis = 1)
```

```python
PCAge = PCA_transform(I, Q)
y = PCAge[0]*1e4
x = freqge[0]*1e-9

plt.plot(x, -y)
plt.title("data after PCA")
plt.show()
```

```python
par_ini = {
    'amp': 1.25,
    'const':0.2,
    'center':7.66,
    'k':0.002
}

par_min = {
    'amp': 0,
    'const':0,
    'center':0,
    'k':0
}

par_max = {
    'amp': np.inf,
    'const':np.inf,
    'center':np.inf,
    'k':np.inf
}

par_vary = {
    'amp': True,
    'const':True,
    'center':True,
    'k':True
}
```

fitting

```python
model = lmfit.Model(Lorentzian)

params = model.make_params()
for name in model.param_names:
    params[name].set(
    value = par_ini[name],
    min = par_min[name],
    max = par_max[name],
    vary = par_vary[name]
    )
```

```python
result = model.fit(data = -y, params = params, freq = x)
```

```python
fig, ax = plt.subplots(figsize=(8,6),constrained_layout=True)
plt.rcParams["font.size"] = 20

plt.tick_params(top='on',bottom='on',left='on',right='on',)

plt.rcParams["xtick.direction"] = 'in'
plt.rcParams["ytick.direction"] = 'in'
result.plot_fit(ax = ax)
# plt.plot(x,result.init_fit)
# plt.xlim(7.95, 8.07)


plt.title("")
plt.legend("")
plt.xlabel("Drive frequency $\omega_\mathrm{d}$ (GHz)")
plt.ylabel("$S_{11}$ (arb. u.)")
# plt.savefig('ge_peak.png', bbox_inches='tight')
plt.show()
```
