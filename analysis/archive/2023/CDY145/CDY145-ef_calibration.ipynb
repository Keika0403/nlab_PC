```python
from plottr.data.datadict_storage import datadict_from_hdf5
import numpy as np
import matplotlib.pyplot as plt
import lmfit
```

```python
import sklearn
from sklearn.decomposition import PCA
```

```python
def Rabi(delay, amp, phi0, detun, const):
    return - amp * np.cos(2*np.pi*detun*delay+phi0) + const
```

```python
header = "D:/K_sunada/database/"
ef = "/2023-07-13/2023-07-13T195105_2ab3d023-td_ef_pi_amp"
dd_ef = datadict_from_hdf5(header + ef +"/data")
```

```python
def PCA_transform(Re_c, Im_c):
    pca = PCA()
    c_data = np.concatenate([np.array([Re_c]).T, np.array([Im_c]).T], axis = 1)
    pca.fit(c_data)
    c_pca = pca.transform(c_data)
    c_PCA1 = c_pca.T[0]
    c_PCA2 = c_pca.T[1]
    return c_PCA1, c_PCA2
```

```python
Rabi_ef = dd_ef['s11']['values']
#imag_ge = Rabi_ge.imag
amp_ef = dd_ef['amplitude']['values']

I = (Rabi_ef.real)
Q = (Rabi_ef.imag)

PCAef = PCA_transform(I, Q)
```

```python
y = PCAef[0]
x = amp_ef

plt.plot(x, y)
plt.show()
```

```python
par_ini = {
    'amp': 0.0001,
    'phi0':np.pi,
    'detun':1/0.3,
    'const':0
}

par_min = {
    'amp': 0,
    'phi0':-2*np.pi,
    'detun':-np.inf,
    'const':0
}

par_max = {
    'amp': np.inf,
    'phi0':2*np.pi,
    'detun':np.inf,
    'const':np.inf,
}

par_vary = {
    'amp': True,
    'phi0':True,
    'detun':True,
    'const':True
}
```

```python
model = lmfit.Model(Rabi)

params = model.make_params()
for name in model.param_names:
    params[name].set(
    value = par_ini[name],
    min = par_min[name],
    max = par_max[name],
    vary = par_vary[name]
    )
```

```python
result_ef = model.fit(data = y, params = params, delay = x)
result_ef
```

```python
fig, ax = plt.subplots(dpi=80)

plt.title("Rabi")

plt.rcParams["xtick.direction"] = 'in'
plt.rcParams["ytick.direction"] = 'in'
result_ef.plot_fit(ax = ax)
plt.plot(amp_ef,result_ef.init_fit)

plt.xlabel("amplitude")
plt.ylabel("mag[S11](a.u.)")


plt.show()
```
